# Analyze

## Load file

```python
import pandas as pd

df_base = pd.read_excel('data/A2_liquicomun/A2_compodem.xlsx', parse_dates=True, index_col=0)
df_base
```

```python
df = df_base.copy()
```

```python
df['cost_eur'] / df['demand_mwh']
```

## Exploratory data analysis

### Overall aggregation

```python
df.pivot_table(index='segment', columns='demand_type', values='price_emwh', aggfunc='mean').style.background_gradient(cmap='Reds', axis=None)
```

### Evolution over time

```python
df = df_base.query('demand_type != "TOTAL"')
df
```

#### Price

```python
import plotly.express as px

px.line(df, x=df.index, y='price_emwh', color='demand_type', facet_col='segment', title='Coste de la energía por MWh', facet_col_wrap=3, height=700)
```

#### Cost

```python
px.area(df, x=df.index, y='cost_eur', color='demand_type', facet_col='segment', title='Coste de la energía por MWh', facet_col_wrap=3, height=700)
```

#### Coste

```python
px.area(df, x=df.index, y='demand_mwh', color='demand_type', facet_col='segment', title='Coste de la energía por MWh', facet_col_wrap=3, height=700)
```
