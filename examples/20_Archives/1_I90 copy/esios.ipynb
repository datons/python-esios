```python
params = {
    "endpoint": "archives/34",  # i90
    "date": "2025-01-01",
    "params_query": {'date_type': 'publicacion'},
}
```

```python
from types import SimpleNamespace as Namespace
p = Namespace(**params)
```

```python
endpoint, i = p.endpoint.split('/')
```

```python
from esios import ESIOSClient

client = ESIOSClient('273695bb4f30b8ab8de054a0410ee2a9d09c97a5006e3a07c27615e6c60888a7')
```

```python
endpoint = client.endpoint(name=endpoint)
series = endpoint.select(id=i)
```

```python
series.configure(date=p.date, **p.params_query)
```

```python
series.__dict__
```

```python
res = series.download_and_extract('data')
```
