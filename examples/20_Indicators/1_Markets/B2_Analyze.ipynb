# Analysis

## Load data

```python
import pandas as pd
df_base = pd.read_excel('markets.xlsx', index_col=0).tz_localize('UTC').tz_convert('Europe/Madrid')
df_base
```

## Visualizations

```python
import plotly.express as px
```

### Raw

#### Line plot

```python
px.line(df_base)
```

#### Area plot

```python
pd.options.plotting.backend = "plotly"
df_base.plot.area(facet_col='variable', facet_col_wrap=3, title='Market evolution')
```

#### Boxplot

```python
df_base.plot.box(color='variable')
```

#### Correlation matrix

```python
df_base.corr().style.background_gradient(cmap='coolwarm')
```

#### Scatter matrix

```python
px.scatter_matrix(df_base)
```

### Aggregated

#### Daily

```python
df = df_base.resample('D').mean()
df.index = df.index.strftime('%Y-%m-%d')
df = df.style.background_gradient(axis=None).format(precision=2)
df
```

#### Hourly

```python
df = df_base.assign(hour = df_base.index.hour)
df = df.groupby('hour').mean()
df
```

```python
df.plot.line()
```
